# __Time-Reversal-symmetry (TRS)__

## _(2D Topological Insulator)_


## Time-Reversal Symmetry and Topological Insulators

### Time-Reversal Symmetry (TRS)

Time reversal is a transformation that reverses the arrow of time:  

$$
\mathcal{T}: t \;\;\longrightarrow\;\; -t
$$

#### Consequences of TRS
- In the presence of a magnetic field, a particle’s trajectory is bent.  
- If we reverse the arrow of time, the particle will not retrace its motion.  
- Therefore, **a magnetic field breaks TR symmetry**.   
- they can exhibit other topological phenomena, such as the **nontrivial $\mathbb{Z}_2$ topological classification**.  


### Chern Insulators vs. TR-Invariant Insulators

- The existence of **chiral edge states** precludes time-reversal symmetry.  
- Time-reversed edge states would describe particles propagating backwards along the edge.  

In **Chern insulators** (two-dimensional insulators with nonvanishing Chern number):  
- The **absence of counterpropagating states** ensures reflectionless propagation along the edges.  
- This follows from the **bulk–boundary correspondence**:  

$$
\text{Net number of edge states at the boundary} = \text{Chern number of the bulk}.
$$



### TR-Invariant Two-Dimensional Insulators

- Time-reversal symmetry ensures that for every edge state mode, there is a **counterpropagating time-reversed partner**.  
- Surprisingly, **no scattering occurs between these two modes**, so reflectionless propagation is still possible.  
- These systems are classified as **two-dimensional topological insulators**.


#### Classification of TR-Invariant 2D Insulators

All two-dimensional TR-invariant insulators fall into two classes:

1. **Trivial class**  
   - Even number of pairs of edge states at a single edge.  

2. **Topological class**  
   - Odd number of pairs of edge states at a single edge.  
   - This is captured by the **$\mathbb{Z}_2$ invariant**.  


### Time Reversal in Quantum Mechanics (Without Spin)

Consider a single particle with no internal degree of freedom, described by a wavefunction \$\psi(\mathbf{r})\$. Its dynamics follow from a time-independent Hamiltonian:

$$
\hat{H} = \big(\hat{\mathbf{p}} - e \mathbf{A}(\mathbf{r})\big)^2 + V(\mathbf{r}),
$$

where \$\mathbf{A}(\mathbf{r})\$ is the vector potential, \$V(\mathbf{r})\$ is the scalar potential, and \$e\$ is the particle’s charge.

The time-dependent Schrödinger equation is:

$$
i \, \partial_t \psi(\mathbf{r}, t) 
= \Big[ \big(-i\nabla - e \mathbf{A}(\mathbf{r})\big)^2 + V(\mathbf{r}) \Big] \psi(\mathbf{r}, t).
$$


#### Complex Conjugation

If \$\psi(\mathbf{r},t)\$ is a solution, then its complex conjugate \$\psi^\*(\mathbf{r},t)\$ satisfies:

$$
- i \, \partial_t \psi^*(\mathbf{r}, t) 
= \Big[ \big(-i\nabla + e \mathbf{A}(\mathbf{r})\big)^2 + V(\mathbf{r}) \Big] \psi^*(\mathbf{r}, t).
$$

This shows that **complex conjugation flips the sign of the vector potential \$\mathbf{A} \to -\mathbf{A}\$**, which corresponds physically to reversing the direction of the magnetic field.


#### Complex Conjugation Operator

We introduce the operator \$\mathcal{K}\$, which acts as complex conjugation in the real-space basis:

$$
\mathcal{K} f(\mathbf{r}) = f^*(\mathbf{r}), 
\qquad \mathcal{K}^2 = 1,
$$

for any complex-valued function \$f(\mathbf{r})\$.

Using \$\mathcal{K}\$, the Schrödinger equation can be rewritten as:

$$
\mathcal{K} \, i \partial_t \psi = \hat{H}^* \, \psi^*.
$$

This formalism sets the stage for understanding **time-reversal symmetry operators** and their consequences when we include **spin** (leading to Kramers’ theorem).


### Time-Reversal Symmetry in Quantum Mechanics

In the special case where the Hamiltonian is **real in the position basis**:

$$
\hat{H}^* = \hat{H},
$$

time-reversal can be implemented simply by acting on the **wavefunction** with the complex conjugation operator \$\mathcal{K}\$.

This happens, for instance, when there is **no vector potential** (\$\mathbf{A} = 0\$). In that case, the Hamiltonian becomes purely real in position space.


#### Time Evolution with Complex Conjugation

Let us consider time evolution for a time \$t\$, then apply the antiunitary operator \$\mathcal{K}\$, then evolve again for time \$t\$, and finally apply \$\mathcal{K}\$ once more:

$$
\mathcal{U} 
= \mathcal{K} e^{-i \hat{H} t} \mathcal{K} e^{-i \hat{H} t}.
$$

Since \$\mathcal{K}\$ conjugates the Hamiltonian,

$$
\mathcal{K} e^{-i \hat{H} t} \mathcal{K} 
= e^{+i \hat{H} t},
$$

we obtain

$$
\mathcal{U} 
= e^{+i \hat{H} t} e^{-i \hat{H} t}.
$$

#### Result

If the Hamiltonian is real (\$\hat{H}^\* = \hat{H}\$), then

$$
\mathcal{U} = 1,
$$

which means that **complex conjugation \$\mathcal{K}\$ acts as the time-reversal operator**.

## Time-Reversal Symmetry and Internal Degrees of Freedom

So far, we discussed time reversal in the case of a single particle with no internal degrees of freedom. However, in **solids** we typically deal with **lattice Hamiltonians**, where:

* The **external degree of freedom** (position) is discrete (lattice sites).
* There are **internal degrees of freedom**, such as:

  * Sublattice index,
  * Orbital type,
  * Spin,
  * or other internal labels.

To account for these, we must generalize the time-reversal operator.


### General Time-Reversal Operator

We define the time-reversal operator \$\mathcal{T}\$ as an **antiunitary** operator of the form

$$
\mathcal{T} = U \, \mathcal{K},
$$

where:

* \$\mathcal{K}\$ is **complex conjugation** in a fixed basis,
* \$U\$ is a **unitary operator** acting only on the internal degrees of freedom.

A Hamiltonian \$\hat{H}\$ is said to be **time-reversal invariant** (TR-invariant) with respect to \$\mathcal{T}\$ if:

$$
\mathcal{T} \, \hat{H} \, \mathcal{T}^{-1} = \hat{H}.
$$

This general definition encompasses both the **scalar case** (\$U = 1\$) and more complicated systems such as electrons with spin.


### Two Possible Types of Time-Reversal Symmetry

Because \$\mathcal{T}\$ is antiunitary, applying it twice gives:

$$
\mathcal{T}^2 = (U \mathcal{K})(U \mathcal{K}) = U U^*.
$$

Here \$U U^\*\$ acts as a unitary operator on the internal degrees of freedom. If this were nontrivial, it would represent an **additional unitary symmetry** of the Hamiltonian. Thus, the only consistent possibility is that

$$
U U^* = e^{i \phi} \, I,
$$

which is equivalent to saying that

$$
\mathcal{T}^2 = e^{i \phi}.
$$

The remarkable result is that the phase factor can only take **two possible values**:

$$
\mathcal{T}^2 = +1 
\quad \text{or} \quad 
\mathcal{T}^2 = -1.
$$


#### Proof Sketch for \$\mathcal{T}^2 = \pm 1\$

Multiplying the relation by \$U^\dagger\$ and iterating, one finds:

$$
U^* = e^{i \phi} U^T,
$$

where the superscript \$T\$ denotes transposition. Repeating this relation again leads to:

$$
U^* = e^{2i\phi} U^*.
$$

Thus \$e^{2i\phi} = 1 ;; \Rightarrow ;; e^{i\phi} = \pm 1\$.

Hence:

$$
\mathcal{T}^2 = \pm 1.
$$


### Examples

* **Case 1: \$\mathcal{T}^2 = +1\$**

  * Example: the scalar time-reversal operator \$\mathcal{T} = \mathcal{K}\$ (complex conjugation only).
  * Appears for spinless particles.

* **Case 2: \$\mathcal{T}^2 = -1\$**

  * Example: electrons with spin-\$\tfrac{1}{2}\$.
  * Time reversal must flip the spin:

    $$
    \mathcal{T} = -i \sigma_y \, \mathcal{K},
    $$

    where \$\sigma\_y\$ is a Pauli matrix.
  * One can check explicitly:

    $$
    \mathcal{T} \, \sigma_j \, \mathcal{T}^{-1} = - \sigma_j, \quad (j = x, y, z).
    $$


### Symmetry of \$U\$

The condition \$\mathcal{T}^2 = \pm 1\$ constrains the unitary matrix \$U\$:

* If \$\mathcal{T}^2 = +1\$:

  $$
  U = U^T \quad \text{(symmetric)}.
  $$

* If \$\mathcal{T}^2 = -1\$:

  $$
  U = - U^T \quad \text{(antisymmetric)}.
  $$

This distinction will become crucial when we analyze the **band structure of TRI insulators**.


### Kramers’ Degeneracy

The most profound consequence of the \$\mathcal{T}^2 = -1\$ case is **Kramers’ theorem**.

Let \$|\psi\rangle\$ be any state. Its time-reversed partner is

$$
|\psi_T\rangle = \mathcal{T} |\psi\rangle.
$$

Because \$\mathcal{T}\$ is antiunitary, we have for any pair of states \$|\psi\rangle, |\phi\rangle\$:

$$
\langle \mathcal{T} \phi | \mathcal{T} \psi \rangle = \langle \phi | \psi \rangle^*.
$$

Now, take \$|\phi\rangle = \mathcal{T}|\psi\rangle\$. Then:

$$
\langle \mathcal{T} \psi | \psi \rangle^* 
= \langle \mathcal{T}^2 \psi | \mathcal{T} \psi \rangle.
$$

* If \$\mathcal{T}^2 = +1\$: this gives no new constraint.
* If \$\mathcal{T}^2 = -1\$: we obtain

$$
\langle \mathcal{T} \psi | \psi \rangle = 0.
$$

Thus, \$|\psi\rangle\$ and \$\mathcal{T}|\psi\rangle\$ are **orthogonal states with the same energy**.

This is **Kramers’ degeneracy**:

* Every eigenstate in a system with \$\mathcal{T}^2 = -1\$ is at least **twofold degenerate**.
* The pair \${ |\psi\rangle, \mathcal{T}|\psi\rangle }\$ is called a **Kramers pair**.


## Time-Reversal Symmetry of a Bulk Hamiltonian

We now analyze the effect of time-reversal symmetry on the **bulk momentum-space Hamiltonian**.


### Plane Wave Basis and Bulk Hamiltonian

For a periodic system (periodic boundary conditions), we work in the plane-wave basis:

$$
|k\rangle = \frac{1}{\sqrt{N_x N_y}} \sum_m e^{i m k} |m\rangle,
$$

where \$|m\rangle\$ labels a lattice site.

Under time reversal \$\mathcal{T} = U \mathcal{K}\$:

$$
\mathcal{T} |k\rangle = |-k\rangle \, \mathcal{T}.
$$

The bulk Hamiltonian can be expressed in this basis as:

$$
\hat{H}_{\text{bulk}} = \sum_k |k\rangle \langle k| \otimes H(k),
$$

where \$H(k)\$ is the **bulk momentum-space Hamiltonian** acting on the internal degrees of freedom.


### Time-Reversal Symmetry Condition

Applying \$\mathcal{T}\$ to the bulk Hamiltonian gives:

$$
\mathcal{T} \, \hat{H}_{\text{bulk}} \, \mathcal{T}^{-1}
= \sum_k |k\rangle \langle k| \otimes \big[ U H(-k)^* U^\dagger \big].
$$

Therefore, **time-reversal invariance** requires:

$$
U \, H(-k)^* \, U^\dagger = H(k).
$$

This is the **fundamental constraint** that TR symmetry imposes on the bulk Hamiltonian.


### Symmetry of the Dispersion Relation

Let \$u\_k\$ be an eigenstate of \$H(k)\$ with energy \$E(k)\$:

$$
H(k) \, |u(k)\rangle = E(k) \, |u(k)\rangle.
$$

Applying the TR condition:

$$
U H(-k)^* U^\dagger |u(k)\rangle = E(k) |u(k)\rangle.
$$

After rearranging, we obtain:

$$
H(-k) \, \big( U^T |u(k)\rangle^* \big) = E(k) \, \big( U^T |u(k)\rangle^* \big).
$$

This shows:

* For every eigenstate at momentum \$k\$, there exists a **time-reversed partner** at momentum \$-k\$ with the **same energy**.
* Thus, the dispersion relation satisfies:

$$
E(k) = E(-k).
$$

> This energy inversion symmetry is **necessary but not sufficient** for full TR symmetry — the internal structure of the eigenstates also matters.


### Time-Reversal Invariant Momenta (TRIM)

A particularly important case occurs at momenta where

$$
k = -k \quad (\text{mod reciprocal lattice vector}),
$$

i.e. where a momentum maps to itself under inversion. These are called **Time-Reversal Invariant Momenta (TRIM)**.

* In \$d\$ dimensions, there are \$2^d\$ TRIM points.
* Example:

  * In 1D → 2 points (\$k = 0, \pi\$).
  * In 2D → 4 points (center and three edges).
  * In 3D → 8 points.

At a TRIM point, the TRS condition reduces to:

$$
U H(k_{\text{TRIM}})^* U^\dagger = H(k_{\text{TRIM}}).
$$


### Kramers’ Degeneracy at TRIM

If \$\mathcal{T}^2 = -1\$ (e.g. electrons with spin-\$\tfrac{1}{2}\$), then **Kramers’ theorem applies**.

* At each TRIM point, every eigenvalue of the Hamiltonian is at least **twofold degenerate**.
* Each state \$|u\rangle\$ has a **Kramers partner** \$\mathcal{T}|u\rangle\$ at the *same momentum* and *same energy*, and they are orthogonal.

This is the microscopic origin of **protected band degeneracies** in TRI systems and a key step toward the \$\mathbb{Z}\_2\$ topological classification.




## Doubling the Hilbert Space for Time-Reversal Symmetry

We want to **construct a Hamiltonian with time-reversal symmetry** starting from a given lattice Hamiltonian \$H\$.


### Step 1. Doubling the System

Take two identical copies of the system:

* Copy 0 carries the original Hamiltonian \$H\$.
* Copy 1 carries its complex-conjugated partner \$H^\* = K H K\$.

We then couple them. The full Hamiltonian is:

$$
\hat{H}_{\text{TRI}} = |0\rangle\langle 0| \otimes H \;+\; |1\rangle\langle 1| \otimes H^*
\;+\;\Big( |0\rangle\langle 1| \otimes I_{\text{ext}} \otimes C \;+\; \text{h.c.} \Big),
$$

where \$C\$ is a coupling operator acting only on the **internal degrees of freedom** (spin, orbital, etc.), and is independent of lattice position.

### Step 2. Matrix Representation

In block-matrix notation, the Hamiltonian looks like:

$$
H_{\text{TRI}} =
\begin{pmatrix}
H & C \\
C^\dagger & H^*
\end{pmatrix}.
$$

Here, the **copy index** is made explicit.


### Step 3. Pauli Operators in Copy Space

We introduce Pauli matrices \$\sigma\_x, \sigma\_y, \sigma\_z\$ acting on the **copy degree of freedom**:

$$
\sigma_{x,y,z} = \sigma_{x,y,z} \otimes I_{\text{ext}} \otimes I_{\text{int}}.
$$

Then we can rewrite:

$$
\hat{H}_{\text{TRI}}
= \frac{1+\sigma_z}{2} \otimes H
+ \frac{1-\sigma_z}{2} \otimes H^*
+ \frac{\sigma_x+i\sigma_y}{2} \otimes C
+ \frac{\sigma_x-i\sigma_y}{2} \otimes C^\dagger.
$$

This makes the time-reversal structure manifest.


### Step 4. Choice of Coupling Operator Determines TR Symmetry

The crucial point: **whether the time-reversal operator squares to \$+1\$ or \$-1\$ depends on the symmetry of \$C\$.**


#### Case A. \$ \mathcal{T}^2 = -1\$ (Kramers TR symmetry)

We choose the TR operator as:

$$
\mathcal{T} = i \sigma_y \, K,
$$

where \$K\$ is complex conjugation.

* Note: \$i\sigma\_y\$ is real, so that the operator acts antiunitarily but keeps things tidy.

Applying this to the Hamiltonian gives:

$$
(i\sigma_y K) H_{\text{TRI}} (i\sigma_y K)^{-1}
= \begin{pmatrix}
H & -C^T \\
-C^* & H^*
\end{pmatrix}.
$$

This equals \$H\_{\text{TRI}}\$ if and only if

$$
C = -C^T,
$$

i.e. \$C\$ must be **antisymmetric**.

So:

* Antisymmetric coupling \$;\Rightarrow; \mathcal{T}^2 = -1\$ (spin-\$\tfrac{1}{2}\$ type TR symmetry).
* This case exhibits **Kramers degeneracy**.


#### Case B. \$ \mathcal{T}^2 = +1\$ (bosonic TR symmetry)

Instead, choose:

$$
\mathcal{T} = \sigma_x K.
$$

Then

$$
(\sigma_x K) H_{\text{TRI}} (\sigma_x K)^{-1}
= \begin{pmatrix}
H & C^T \\
C^* & H^*
\end{pmatrix}.
$$

This equals \$H\_{\text{TRI}}\$ if and only if

$$
C = C^T,
$$

i.e. \$C\$ must be **symmetric**.

So:

* Symmetric coupling \$;\Rightarrow; \mathcal{T}^2 = +1\$.


### Step 5. Special Case: Only Want \$ \mathcal{T}^2 = +1\$

If we only want TRS with \$\mathcal{T}^2 = +1\$, we don’t even need to double the Hilbert space.

* Simply define

$$
H_{\text{TRI}} = \frac{1}{2}\big( H + H^* \big).
$$

This is equivalent to taking the **real part** of the Hamiltonian.


### Summary

* **Doubling trick**: build a TR-invariant Hamiltonian by combining \$H\$ and \$H^\*\$.
* **Coupling operator \$C\$** decides the symmetry class:

  * Antisymmetric \$C ; \Rightarrow; \mathcal{T}^2 = -1\$ (fermionic TR, Kramers degeneracy).
  * Symmetric \$C ; \Rightarrow; \mathcal{T}^2 = +1\$ (bosonic TR).
* For \$\mathcal{T}^2 = +1\$, doubling isn’t necessary: just take the real part.
