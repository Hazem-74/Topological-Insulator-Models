# __Tight Binding: Linear Combination of Atomic Orbitals__


## Deriving the Tight-Binding Model from First Principles

So far, we have simply introduced the tight-binding model as a phenomenological description of electrons hopping between localized atomic orbitals. However, it is insightful to ask whether this model can be derived from more fundamental principles. Doing so reveals both the physical intuition behind the model and its limitations.

We begin with a single isolated atom placed at the origin. The Hamiltonian for an electron in this atom takes the form:

$$
H_{\text{atom}} = \frac{p^2}{2m} + V_{\text{atom}}(x)
$$

The electron is bound to the atom in discrete energy levels $ \epsilon_n < 0 $, with corresponding eigenstates $ \phi_n(x) $ satisfying:

$$
H_{\text{atom}} \phi_n(x) = \epsilon_n \phi_n(x)
$$

In addition to these bound states, there also exist scattering states with $ \epsilon > 0 $, which are not localized.

We now consider a periodic lattice of such atoms. The resulting total potential is constructed as a sum over translated copies of the atomic potential:

$$
V_{\text{lattice}}(x) = \sum_{r \in \Lambda} V_{\text{atom}}(x - r)
$$

Here $ \Lambda $ denotes the set of lattice sites. In the presence of this periodic potential, the atomic energy levels are modified. Electrons with deep binding energies remain largely localized near their original atoms. However, those with weaker binding become delocalized due to the overlap of their wavefunction tails with neighboring atoms. These weakly bound electrons, often called *valence electrons*, are responsible for forming energy bands and giving rise to electrical conductivity.

This physical picture is captured by the tight-binding model. To formalize this, we attempt to write the eigenstates of the full lattice Hamiltonian

$$
H = \frac{p^2}{2m} + V_{\text{lattice}}(x)
$$

as linear combinations of the localized atomic orbitals. A natural ansatz for a Bloch state is:

$$
\psi_k(x) = \frac{1}{\sqrt{N}} \sum_{r \in \Lambda} e^{i k \cdot r} \phi(x - r)
$$

This function satisfies Bloch's condition $ \psi_k(x + a) = e^{i k \cdot a} \psi_k(x) $ for any lattice vector $ a \in \Lambda $, as expected for states in a periodic potential. Unlike Wannier functions, the orbitals $ \phi(x - r) $ centered at different sites are not generally orthogonal, which must be accounted for in the derivation.
a
The energy expectation value for such a state is given by:

$$
E(k) = \frac{\langle \psi_k | H | \psi_k \rangle}{\langle \psi_k | \psi_k \rangle}
$$

The normalization becomes:

$$
\langle \psi_k | \psi_k \rangle = \sum_{r} e^{-i k \cdot r} \alpha(r)
$$

where $ \alpha(r) $ represents the overlap between wavefunctions centered at different sites:

$$
\alpha(r) = \int d^3x \, \phi^*(x) \phi(x + r)
$$

For the numerator, we express the full Hamiltonian as $ H = H_{\text{atom}} + \Delta V(x) $, where

$$
\Delta V(x) = V_{\text{lattice}}(x) - V_{\text{atom}}(x)
$$

Then, the expectation value becomes:

$$
\langle \psi_k | H | \psi_k \rangle = \epsilon \langle \psi_k | \psi_k \rangle + \Delta \epsilon + \sum_{r \ne 0} e^{-i k \cdot r} \gamma(r)
$$

Here $ \Delta \epsilon $ captures the shift in the atomic energy due to the presence of neighboring potentials:

$$
\Delta \epsilon = \int d^3x \, \phi^*(x) \Delta V(x) \phi(x)
$$

and $ \gamma(r) $ represents the overlap between the potential at one site and a neighboring wavefunction:

$$
\gamma(r) = \int d^3x \, \phi^*(x - r) \Delta V(x) \phi(x)
$$

Combining everything, the energy dispersion relation becomes:

$$
E(k) = \epsilon + \Delta \epsilon + \frac{\sum_{r \ne 0} e^{-i k \cdot r} \left[ \gamma(r) - \alpha(r) \Delta \epsilon \right]}{1 + \sum_{r \ne 0} e^{-i k \cdot r} \alpha(r)}
$$

Assuming $ \alpha(r) \ll 1 $, we can expand the denominator and simplify the result to:

$$
E(k) \approx \epsilon + \Delta \epsilon + \sum_{r \ne 0} e^{-i k \cdot r} \left[ \gamma(r) - \alpha(r) \Delta \epsilon \right]
$$

This approximation forms the basis of the tight-binding model. It captures the essential physics of band formation due to electron hopping between atomic orbitals. In practical applications, we often retain only the nearest-neighbor contributions to $ \gamma(r) $ and $ \alpha(r) $, since these overlap integrals decay rapidly with distance.


## LCAO:

To describe such delocalized states, we use the **Linear Combination of Atomic Orbitals (LCAO)** ansatz:
$$
\psi_k(x) = \frac{1}{\sqrt{N}} \sum_{r \in \Lambda} e^{i k \cdot r} \sum_n c_n \phi_n(x - r)
$$
where:
- $ k $ is the crystal momentum,
- $ \phi_n(x - r) $ is a localized orbital (e.g., $ s $, $ p $, $ d $) centered at site $ r $,
- $ c_n $ are complex coefficients to be determined,
- $ N $ is the number of lattice sites (normalization factor).

This wavefunction satisfies **Bloch’s theorem**, $ \psi_k(x + a) = e^{i k \cdot a} \psi_k(x) $, and represents a **Bloch state** constructed from atomic-like orbitals.

### Energy Expectation Value

The energy dispersion $ E(k) $ is obtained variationally:
$$
E(k) = \frac{\langle \psi_k | H | \psi_k \rangle}{\langle \psi_k | \psi_k \rangle}
$$
where $ H = \frac{p^2}{2m} + V_{\text{lattice}}(x) $. We decompose $ H = H_{\text{atom}} + \Delta V(x) $, with $ \Delta V(x) = V_{\text{lattice}}(x) - V_{\text{atom}}(x) $ representing the perturbation due to neighboring atoms.

We define two key integrals:

- **Overlap integral**:
  $$
  \alpha_{n,n'}(r) = \int d^3x\, \phi_n^*(x - r) \phi_{n'}(x)
  $$
  This measures the spatial overlap between orbitals on different sites.

- **Hamiltonian matrix element**:
  $$
  \gamma_{n,n'}(r) = \int d^3x\, \phi_n^*(x - r) \left(H_{\text{atom}} + \Delta V\right) \phi_{n'}(x)
  $$

Using these, the normalization and energy expectation become:
$$
\langle \psi_k | \psi_k \rangle = \sum_{r \in \Lambda} \sum_{n,n'} c_n^* c_{n'} e^{-i k \cdot r} \alpha_{n,n'}(r)
$$

$$
\langle \psi_k | H | \psi_k \rangle = \sum_{r \in \Lambda} \sum_{n,n'} c_n^* c_{n'} e^{-i k \cdot r} \left[ \epsilon_n \alpha_{n,n'}(r) + \gamma_{n,n'}(r) \right]
$$

Note: The sums include $ r = 0 $, so $ \alpha_{n,n'}(0) = \delta_{nn'} $ (if orbitals are orthonormal), and $ \gamma_{n,n}(0) = \epsilon_n $.

### Variational Principle and the Secular Equation

Minimizing $ E(k) $ with respect to $ c_n^* $ leads to:
$$
\frac{\partial}{\partial c_n^*} \langle \psi_k | H | \psi_k \rangle = E(k) \frac{\partial}{\partial c_n^*} \langle \psi_k | \psi_k \rangle
$$
This results in a **generalized eigenvalue problem**:
$$
\sum_{n'} M_{n,n'}(k) \, c_{n'} = 0
$$
where $ M_{n,n'}(k) $ is the Hermitian matrix:
$$
M_{n,n'}(k) = \sum_{r \in \Lambda} e^{-i k \cdot r} \left[ \gamma_{n,n'}(r) - (E(k) - \epsilon_n) \alpha_{n,n'}(r) \right]
$$

For non-trivial solutions, we require:
$$
\det \left[ M_{n,n'}(k) \right] = 0
$$

### Solving for the Band Structure

Suppose we include $ p $ atomic orbitals per site (e.g., $ s, p_x, p_y, p_z $). Then $ M_{n,n'}(k) $ is a $ p \times p $ matrix, and $ \det M = 0 $ defines a polynomial equation in $ E(k) $ of degree $ p $. Its $ p $ roots give the **energy bands** $ E_1(k), E_2(k), \dots, E_p(k) $.

For each band:
- The energy $ E_n(k) $ defines the dispersion relation.
- The null eigenvector $ c_n $ reveals how atomic orbitals mix to form the Bloch state.

### Connection to the Simple Tight-Binding Model

In many cases:
- Overlap $ \alpha_{n,n'}(r) \ll 1 $ for $ r \neq 0 $,
- $ \gamma_{n,n'}(r) $ decays rapidly with distance.

Then we approximate:
- $ \alpha_{n,n'}(r) \approx \delta_{n,n'} \delta_{r,0} $,
- Keep only nearest-neighbor $ \gamma_{n,n'}(r) $.

This reduces the full LCAO result to the familiar tight-binding form:
$$
H = \sum_r \epsilon_n |r,n\rangle\langle r,n| - \sum_{\langle r,r'\rangle} t_{n,n'} \left( |r,n\rangle\langle r',n'| + \text{h.c.} \right)
$$
with hopping $ t_{n,n'} \propto \gamma_{n,n'}(a) $.


## **S-orbital:**

- Assume that the overlap integrals $ \alpha(r) $ and coupling terms $ \gamma(r) $ are significant **only between nearest-neighbor sites** — a reasonable assumption since atomic wavefunctions decay rapidly with distance. All contributions from more distant sites are neglected.

Furthermore, suppose the valence electron occupies an **s-orbital**. This has two important consequences:
1. The orbital is **spherically symmetric**: $ \phi(\mathbf{x} - r) = \phi(|\mathbf{x} - r|) $,
2. The wavefunction can be taken **real-valued**: $ \phi^*(x) = \phi(x) $.

Under these conditions, the overlap integral becomes:
$$
\alpha(r) = \int d^3x\, \phi(x - r)\, \phi(x) = \alpha(-r)
$$
which is symmetric under reversal of the displacement vector.

To obtain a similar symmetry for $ \gamma(r) $, we assume the crystal has **inversion symmetry**, i.e., $ \Delta V(x) = \Delta V(-x) $. Then:
$$
\gamma(r) = \int d^3x\, \phi(x - r)\, \Delta V(x)\, \phi(x)
$$
Changing variables to $ x' = -x $, and using both inversion symmetry and the spherical symmetry of the s-orbital:
$$
\gamma(r) = \int d^3x'\, \phi(-x' - r)\, \Delta V(-x')\, \phi(-x') = \int d^3x'\, \phi(|x' + r|)\, \Delta V(x')\, \phi(|x'|)
= \gamma(-r)
$$
Thus, $ \gamma(r) $ is also symmetric: $ \gamma(r) = \gamma(-r) $.

Now recall the general dispersion relation derived earlier:
$$
E(k) = \epsilon + \Delta\epsilon + \frac{\sum_{r \ne 0} e^{-i k \cdot r} \left[ \gamma(r) - \Delta\epsilon\, \alpha(r) \right]}{1 + \sum_{r \ne 0} e^{-i k \cdot r} \alpha(r)}
$$

With nearest-neighbor approximation and $ \alpha(r) \ll 1 $, we expand the denominator to linear order:
$$
E(k) \approx \epsilon + \Delta\epsilon + \sum_{r \ne 0} e^{-i k \cdot r} \left[ \gamma(r) - \Delta\epsilon\, \alpha(r) \right]
$$

Because the lattice is symmetric under $ r \to -r $, and both $ \alpha(r) $ and $ \gamma(r) $ are even functions, we can pair each site $ r $ with $ -r $. Then:
$$
\sum_{r} e^{-i k \cdot r} \left[ \gamma(r) - \Delta\epsilon\, \alpha(r) \right] = \sum_{a} 2 \cos(k \cdot a) \left[ \gamma(a) - \Delta\epsilon\, \alpha(a) \right]
$$
where the sum is now over **distinct nearest-neighbor vectors** $ a $ from a given site.

Thus, the energy dispersion becomes:
$$
E(k) = \epsilon + \Delta\epsilon + 2 \sum_{a} \cos(k \cdot a) \left[ \gamma(a) - \Delta\epsilon\, \alpha(a) \right]
$$

This is precisely the form of the tight-binding dispersion we derived phenomenologically earlier.

Comparing with the original tight-binding result:
$$
E(k) = E_0 - 2t \sum_a \cos(k \cdot a)
$$
we identify:
- The on-site energy: $ E_0 = \epsilon + \Delta\epsilon $
- The hopping amplitude: $ t_a = -\left[ \gamma(a) - \Delta\epsilon\, \alpha(a) \right] $

Hence, the **hopping parameter $ t $** is not a free parameter, but arises from fundamental quantities:
- $ \gamma(a) $: the inter-site potential coupling,
- $ \alpha(a) $: the orbital overlap,
- $ \Delta\epsilon $: the energy shift due to the crystalline environment.


## **Ashcroft Example:**

### Applying the Tight-Binding Model to a Face-Centered Cubic (FCC) Crystal

The tight-binding dispersion relation for an electron in a periodic lattice is given by:

$$
\varepsilon(\mathbf{k}) = E_s - \beta - \sum_{\text{n.n.}} \gamma(\mathbf{R}) \cos(\mathbf{k} \cdot \mathbf{R}),
$$

* where the sum runs over all nearest-neighbor vectors $ \mathbf{R} $ connecting a site to its adjacent neighbors in the Bravais lattice.

* **On-site energy** $E_s$:
  This corresponds to the energy of an electron in an isolated atomic orbital, including the central atomic potential $U(\mathbf{r})$:

  $$
  E_s = \int d^3\mathbf{r} \, \phi(\mathbf{r}) \, H_{\text{atom}} \, \phi(\mathbf{r})
  = \int d^3\mathbf{r} \, \phi(\mathbf{r}) \left[ -\frac{\hbar^2}{2m} \nabla^2 + U(\mathbf{r}) \right] \phi(\mathbf{r})
  $$

* **Correction due to lattice potential** $\beta$:
  This represents the interaction of the atomic orbital at one site with the periodic **crystal potential** centered at the **same site** (arising from the rest of the lattice):

  $$
  \beta = \int d^3\mathbf{r} \, \phi(\mathbf{r}) \, \Delta U(\mathbf{r}) \, \phi(\mathbf{r})
  $$

  where $\Delta U(\mathbf{r}) = U_{\text{crystal}}(\mathbf{r}) - U_{\text{atom}}(\mathbf{r})$ captures the difference between the full periodic crystal potential and the atomic potential.

#### Nearest Neighbors in the FCC Lattice

For a face-centered cubic (FCC) lattice, each atom has **12 nearest neighbors**. These are located at:

$$
\mathbf{R} = \frac{a}{2} (\pm 1, \pm 1, 0), \quad \frac{a}{2} (\pm 1, 0, \pm 1), \quad \frac{a}{2} (0, \pm 1, \pm 1),
$$

where $ a $ is the cubic lattice constant. These vectors point to the centers of the faces of the cube and define the local coordination environment.

As shown in the figure below, these 12 points form a symmetric shell around the origin:

![Figure: Nearest Neighbors in FCC Lattice](FCC.png)

#### Dot Products with Crystal Momentum

Let $ \mathbf{k} = (k_x, k_y, k_z) $. The scalar products $ \mathbf{k} \cdot \mathbf{R} $ take the form:

$$
\mathbf{k} \cdot \mathbf{R} = \frac{a}{2} (\pm k_x \pm k_y), \quad \frac{a}{2} (\pm k_y \pm k_z), \quad \frac{a}{2} (\pm k_z \pm k_x).
$$

Each combination appears once, and due to the cosine function’s even symmetry, $ \cos(-\theta) = \cos(\theta) $, contributions from opposite directions add constructively.

#### Symmetry and Uniform Hopping

We assume:
- The atomic potential has **cubic symmetry**: $ \Delta U(x, y, z) = \Delta U(\text{perm}, \text{sign flips}) $,
- The valence orbital is an **s-orbital**, which is spherically symmetric and real: $ \phi(\mathbf{r}) = \phi(r) $.

These conditions imply that the hopping integral $ \gamma(\mathbf{R}) $ is identical for all 12 nearest-neighbor vectors. We denote this common value as $ \gamma $:

$$
\gamma = -\int d^3\mathbf{r} \, \phi(\mathbf{r}) \, \Delta U(\mathbf{r}) \, \phi(\mathbf{r} - \mathbf{R}),
$$

where $ |\mathbf{R}| = a/\sqrt{2} $. This integral determines the strength of inter-site coupling and is typically negative or small and positive depending on the orbital overlap and potential shape.

#### Final Dispersion Relation

Using the symmetry and uniformity of $ \gamma $, the sum over $ \cos(\mathbf{k} \cdot \mathbf{R}) $ simplifies. Grouping terms by coordinate pairs, we find:

$$
\varepsilon(\mathbf{k}) = E_s - \beta - 4\gamma \left[ 
\cos\left(\frac{k_x a}{2}\right)\cos\left(\frac{k_y a}{2}\right) +
\cos\left(\frac{k_y a}{2}\right)\cos\left(\frac{k_z a}{2}\right) +
\cos\left(\frac{k_z a}{2}\right)\cos\left(\frac{k_x a}{2}\right)
\right]
$$

This is the **tight-binding energy band** for an s-orbital on an FCC lattice with nearest-neighbor hopping.


```python
# Main Libraries
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from scipy.stats import gaussian_kde
import plotly.graph_objects as go
```

```python
# Parameters
a = 1.0
gamma = 0.8
Es = 0.0
beta = 0.0
n_points = 100
```

```python
# FCC Dispersion Function
def fcc_dispersion(kx, ky, kz, a=1.0, gamma=1.0, Es=0.0, beta=0.0):
    term1 = np.cos(kx * a / 2) * np.cos(ky * a / 2)
    term2 = np.cos(ky * a / 2) * np.cos(kz * a / 2)
    term3 = np.cos(kz * a / 2) * np.cos(kx * a / 2)
    return Es - beta - 4 * gamma * (term1 + term2 + term3)

# Define grid and evaluate dispersion
kx = np.linspace(-np.pi, np.pi, n_points)
ky = np.linspace(-np.pi, np.pi, n_points)
kz = np.linspace(-np.pi, np.pi, n_points)

KX, KY = np.meshgrid(kx, ky)
```

```python
# Complute Energy
E_slice = fcc_dispersion(KX, KY, kz=0, a=a, gamma=gamma, Es=Es, beta=beta)

# Plotting
fig = plt.figure(figsize=(12, 5))
ax3d = fig.add_subplot(121, projection='3d')

ax3d.plot_surface(KX, KY, E_slice, cmap='viridis', edgecolor='none')
ax3d.set_xlabel('$k_x$')
ax3d.set_ylabel('$k_y$')
ax3d.set_zlabel(r'$\epsilon(k_x, k_y)$')
ax3d.set_title('FCC Tight-Binding Dispersion ($k_z=0$)')
```

```python
fig = plt.figure(figsize=(18, 6))

# Contour Plot
ax1 = fig.add_subplot(1, 2, 1)
contour = ax1.contour(KX, KY, E_slice, levels=20, cmap='magma')
ax1.set_title('2D Contour of $E(k_x, k_y)$')
ax1.set_xlabel('$k_x$')
ax1.set_ylabel('$k_y$')
fig.colorbar(contour, ax=ax1, label='Energy [eV]')

# Filled Contour Plot
ax2 = fig.add_subplot(1, 2, 2)
contourf = ax2.contourf(KX, KY, E_slice, levels=30, cmap='magma_r')
ax2.set_title('Filled Contour of $E(k_x, k_y)$')
ax2.set_xlabel('$k_x$')
ax2.set_ylabel('$k_y$')
fig.colorbar(contourf, ax=ax2, label='Energy [eV]')

plt.tight_layout()
plt.show()
```

```python
# Flatten the energy grid for DOS
E_flat = E_slice.flatten()

# Perform KDE
kde = gaussian_kde(E_flat)
E_vals = np.linspace(E_flat.min(), E_flat.max(), 1000)
kde_vals = kde(E_vals)

# Plot both Histogram and KDE in subplots
fig, axs = plt.subplots(1, 2, figsize=(12, 4), sharey=True)

# Histogram
axs[0].hist(E_flat, bins=100, density=True, color='steelblue', alpha=0.8)
axs[0].set_title('DOS Histogram')
axs[0].set_xlabel('Energy (eV)')
axs[0].set_ylabel('DOS (arb. units)')
axs[0].grid(True)

# KDE Plot
axs[1].plot(E_vals, kde_vals, color='steelblue', lw=2)
axs[1].fill_between(E_vals, kde_vals, color='lightblue', alpha=0.4)
axs[1].set_title('DOS KDE')
axs[1].set_xlabel('Energy (eV)')
axs[1].grid(True)

plt.tight_layout()
plt.show()
```

```python
#  3D k-space grid 
kx_vals = np.linspace(-np.pi, np.pi, n_points)
ky_vals = np.linspace(-np.pi, np.pi, n_points)
kz_vals = np.linspace(-np.pi, np.pi, n_points)
KX, KY, KZ = np.meshgrid(kx_vals, ky_vals, kz_vals, indexing='ij')

#  3D dispersion function 
E_3D = fcc_dispersion(KX, KY, KZ)
```

```python
# === Extract Fermi Surface Points ===
E_fermi = -1.5
fermi_tol = 0.06  # Energy window around E_F
fermi_mask = np.abs(E_3D - E_fermi) < fermi_tol

kx_fs = KX[fermi_mask]
ky_fs = KY[fermi_mask]
kz_fs = KZ[fermi_mask]

# === Plot Fermi Surface as 3D Scatter ===
fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')

ax.scatter(kx_fs, ky_fs, kz_fs, c='crimson', s=2, alpha=0.6, label='Fermi Surface')

ax.set_title("Fermi Surface in 3D Tight-Binding Model")
ax.set_xlabel('$k_x$')
ax.set_ylabel('$k_y$')
ax.set_zlabel('$k_z$')
ax.set_xlim(-np.pi, np.pi)
ax.set_ylim(-np.pi, np.pi)
ax.set_zlim(-np.pi, np.pi)
ax.legend()

plt.tight_layout()
plt.show()
```

```python
# Create Interactive 3D Scatter Plot
fig = go.Figure(data=[go.Scatter3d(
    x=kx_fs, y=ky_fs, z=kz_fs,
    mode='markers',
    marker=dict(size=3, color='crimson', opacity=0.3),
    name='Fermi Surface'
)])

fig.update_layout(
    title="Fermi Surface in 3D Tight-Binding Model",
    scene=dict(
        xaxis_title='k_x',
        yaxis_title='k_y',
        zaxis_title='k_z',
        xaxis_range=[-np.pi, np.pi],
        yaxis_range=[-np.pi, np.pi],
        zaxis_range=[-np.pi, np.pi],
        aspectmode='cube'
    ),
    width=800,
    height=600,
    margin=dict(l=0, r=0, b=0, t=40)
)

# Show Interactive Plot
fig.show()
```
