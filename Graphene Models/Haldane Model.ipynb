# __Haldane Model__

## _(2D Topological Insulator)_


The **Haldane model** is for realizing a **Chern insulator** (or **quantum anomalous Hall effect**) without an external magnetic field as it describes spinless fermions hopping on a **honeycomb lattice** with:  

- **Nearest-neighbor hopping**: generates graphene-like Dirac cones.  
- **Next-nearest-neighbor hopping with complex phases**: breaks time-reversal symmetry without net magnetic flux through the unit cell.  
- **Sublattice potential difference (mass term)**: breaks inversion symmetry.  

This interplay leads to a nontrivial **bulk Chern number**, giving rise to **topologically protected edge states** via the **bulk–boundary correspondence**.

### Hamiltonian (from Haldane Paper):

The tight-binding Hamiltonian of the Haldane model can be written as

$$
\hat{H} = 
-t_1 \sum_{\langle i,j \rangle} 
   \left( c_i^\dagger c_j + \text{h.c.} \right)
- t_2 \sum_{\langle\!\langle i,j \rangle\!\rangle} 
   \left( e^{i \nu_{ij} \phi} c_i^\dagger c_j + \text{h.c.} \right)
+ t_0 \sum_i \xi_i \, c_i^\dagger c_i
$$

where:

- $ c_i^\dagger, c_i $ are fermionic creation/annihilation operators.  
- $ t_1 \equiv t $: nearest-neighbor hopping amplitude.  
- $ t_2 $: next-nearest-neighbor hopping amplitude.  
- $ \phi $: complex hopping phase (mimics orbital magnetic effects without net flux).  
- $ t_0 \equiv t_0 $: on-site energy imbalance (staggered sublattice potential).  
- $ \nu_{ij} = \pm 1$: orientation-dependent phase factor.  
- $ \xi_i = \pm 1$: sublattice index ($+1$ for A, $-1$ for B). 

### Hamiltonian in Dirac notation:

$$
\boxed{
\hat{H} =  t_1 \sum_{\langle n_\alpha, m_\beta \rangle} 
\left( |n_\alpha \rangle\langle m_\beta| + |m_\beta\rangle\langle n_\alpha| \right) 
+ t_2 \sum_{\langle\!\langle n_\alpha, m_\alpha \rangle\!\rangle}  
\left( e^{i \nu_{n_\alpha m_\alpha} \phi} |n_\alpha\rangle\langle m_\alpha| + e^{-i \nu_{n_\alpha m_\alpha} \phi} |m_\alpha\rangle\langle n_\alpha| \right) 
+ t_0 \sum_{n_\alpha} \xi_\alpha |n_\alpha\rangle\langle n_\alpha|
}
$$

 Where:

* $n, m \in \mathbb{Z}^2$: integers labeling the unit cell position.
* $\alpha, \beta \in \{A, B\}$: sublattice index.
* $n_A$: A-site in cell $n$.
* $m_B$: B-site in cell $m$.
* $\langle n_\alpha, m_\beta \rangle$: nearest-neighbor (NN) pairs.
* $\langle\!\langle n_\alpha, m_\alpha \rangle\!\rangle$: next-nearest-neighbor (NNN) pairs within the same sublattice.
* $t_1$: nearest-neighbor hopping amplitude.
* $t_2 e^{\pm i \phi}$: complex next-nearest-neighbor hopping, with phase $\phi$.
* $\nu_{n_\alpha m_\alpha} = \pm 1$ depending on whether the hopping path is clockwise or counterclockwise.
* $t_0$: sublattice mass term (on-site potential).
* $\xi_\alpha = +1$ for $\alpha=A$, $-1$ for $\alpha=B$.

### Momentum-Space Representation

the Hamiltonian in the **sublattice (pseudospin) basis** becomes:

$$
H(\mathbf{k}) = d_x(\mathbf{k}) \, \sigma_x \;+\; d_y(\mathbf{k}) \, \sigma_y \;+\; d_z(\mathbf{k}) \, \sigma_z
$$

with components:

$$
\begin{aligned}
d_x(\mathbf{k}) &= -t \sum_{j=1}^3 \cos(\mathbf{k}\cdot \mathbf{a}_j), \\
d_y(\mathbf{k}) &= -t \sum_{j=1}^3 \sin(\mathbf{k}\cdot \mathbf{a}_j), \\
d_z(\mathbf{k}) &= t_0 - 2t_2 \sin \phi \sum_{j=1}^3 \sin(\mathbf{k}\cdot \mathbf{b}_j).
\end{aligned}
$$

Here:
- $ \mathbf{a}_j $: nearest-neighbor vectors: 
 $$
 \mathbf{a}_1 = \left(0, - \textbf{a} \right), \quad 
 \mathbf{a}_2 = \left( \tfrac{\sqrt{3}}{2}\textbf{a}, \tfrac{1}{2}\textbf{a} \right), \quad
 \mathbf{a}_3 = \left( -\tfrac{\sqrt{3}}{2}\textbf{a}, \tfrac{1}{2}\textbf{a} \right).
 $$

- $ \mathbf{b}_j $: next-nearest-neighbor vectors:

$$
\mathbf{b}_1 = \left( \sqrt{3}\textbf{a}, \, 0 \right), \\
\mathbf{b}_2 = \left( -\tfrac{\sqrt{3}}{2}\textbf{a}, \, -\tfrac{3}{2}\textbf{a} \right), \\
\mathbf{b}_3 = \left( -\tfrac{\sqrt{3}}{2}\textbf{a}, \, \tfrac{3}{2}\textbf{a} \right).
$$

- Here $\textbf{a}$ is the lattice nearest-neighbor distance on the honeycomb lattice.
- $ \sigma_{x,y,z} $: Pauli matrices. 

### Bulk Dispersion Relation

The eigenvalues of the Bloch Hamiltonian are

$$
E_{\pm}(\mathbf{k}) = \pm \sqrt{d_x(\mathbf{k})^2 + d_y(\mathbf{k})^2 + d_z(\mathbf{k})^2}
$$

describing conduction ($+$) and valence ($-$) bands.



### Eigenstates Shape (Pseudospinors)

The eigenstates can be written as **pseudospinors**:

$$
|\psi_\pm(\mathbf{k})\rangle 
= \frac{1}{\sqrt{2}}
\begin{pmatrix}
\sqrt{1 \pm \dfrac{d_z(\mathbf{k})}{|d(\mathbf{k})|}} \\
\frac{d_x(\mathbf{k}) + i d_y(\mathbf{k})}{\sqrt{d_x(\mathbf{k})^2 + d_y(\mathbf{k})^2}} \, 
\sqrt{1 \mp \dfrac{d_z(\mathbf{k})}{|d(\mathbf{k})|}}
\end{pmatrix}
$$

where $ |d(\mathbf{k})| = \sqrt{d_x^2 + d_y^2 + d_z^2} $.


### Gap Closing Conditions

The bulk gap closes precisely when 
$$
d_x = d_y = d_z = 0
$$
at some $\mathbf{k}$ in the Brillouin zone (BZ).  
The only possible gap-closing points are the Dirac points $K$ and $K'$, given by
$$
K = \left( \tfrac{4\pi}{3\sqrt{3}\textbf{a}},\, 0 \right), 
\qquad
K' = \left( -\tfrac{4\pi}{3\sqrt{3}\textbf{a}},\, 0 \right).
$$

At $K$ one finds:
$$
d_x(K) = 0, \qquad d_y(K) = 0, \qquad d_z(K) = t_0 - 3\sqrt{3}\,t_2 \sin\phi ,
$$
and similarly at $K'$:
$$
d_x(K') = 0, \qquad d_y(K') = 0, \qquad d_z(K') = t_0 + 3\sqrt{3}\,t_2 \sin\phi .
$$

Therefore, the gap closes when
$$
d_z(K) = 0 \;\;\;\;\Longrightarrow\;\;\;\; t_0 = 3\sqrt{3}\,t_2 \sin\phi ,
$$
or
$$
d_z(K') = 0 \;\;\;\;\Longrightarrow\;\;\;\; t_0 = -3\sqrt{3}\,t_2 \sin\phi .
$$


### Dirac Points and Cones

- The honeycomb lattice naturally hosts **Dirac cones** at $K$ and $K'$.  
- The staggered mass term $t_0$ and complex hopping phase $\phi$ determine whether these cones open into a trivial or topological gap.  
- Changing the sign of the effective mass at the Dirac points signals a **topological phase transition**.



### Bulk–Boundary Correspondence

- If the **bulk Chern number** $Q$ is nonzero, the system supports **chiral edge states** at open boundaries.  
- The number of edge states equals the bulk Chern number.  
- These edge states are **topologically protected** against backscattering, realizing the **anomalous quantum Hall effect** without Landau levels.



```python
# Main Library
import sympy as sp
import numpy as np
from joblib import Parallel, delayed
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
import plotly.graph_objects as go
from IPython.display import Image
from mpl_toolkits.mplot3d import Axes3D
from sympy import matrix2numpy
```

```python
# Define symbols
kx, ky, t1, t2, t0, phi = sp.symbols('k_x k_y t1 t2 t0 phi', real=True)

# Define lattice constant
a = sp.symbols('a', real=True, positive=True)

# Pauli matrices
sigma_x = sp.Matrix([[0, 1],[1, 0]])
sigma_y = sp.Matrix([[0, -sp.I],[sp.I, 0]])
sigma_z = sp.Matrix([[1, 0],[0, -1]])

# Nearest-neighbor vectors
a1 = sp.Matrix([0, -a])
a2 = sp.Matrix([sp.sqrt(3)/2 * a, sp.Rational(1,2) * a])
a3 = sp.Matrix([-sp.sqrt(3)/2 * a, sp.Rational(1,2) * a])
a_vecs = [a1, a2, a3]

# Nearest-neighbor hopping terms (for d_x and d_y)
d_x = -t1 * sum(sp.cos(kx*v[0] + ky*v[1]) for v in a_vecs)
d_y = -t1 * sum(sp.sin(kx*v[0] + ky*v[1]) for v in a_vecs)

# Next-nearest neighbor vectors (NNN)
b1 = sp.Matrix([sp.sqrt(3) * a, 0])
b2 = sp.Matrix([-sp.sqrt(3)/2 * a, -sp.Rational(3,2) * a])
b3 = sp.Matrix([-sp.sqrt(3)/2 * a,  sp.Rational(3,2) * a])
b_vecs = [b1, b2, b3]

# d_z term (Semenoff mass + Haldane phase contribution)
d_z = t0 - 2 * t2 * sp.sin(phi) * sum(sp.sin(kx*v[0] + ky*v[1]) for v in b_vecs)
```

```python
# Hamiltonian in 2x2 form
H = d_x*sigma_x + d_y*sigma_y + d_z*sigma_z
H = sp.simplify(H)

# Display the Hamiltonion
H
```

```python
'''Changing to numpy rather sympy'''
# Parameters
t1 = 1.0        # Nearest-neighbor hopping
t2 = 0.3        # Next-nearest-neighbor hopping
t0 = 0.2        # Staggered potential 
phi = np.pi / 2 # Haldane phase
a = 1.0         # Lattice constant 

# Nearest-Neighbor Vectors (from A to B)
delta1 = np.array([0.0, -a])
delta2 = np.array([np.sqrt(3)/2 * a, 0.5 * a])
delta3 = np.array([-np.sqrt(3)/2 * a, 0.5 * a])
deltas = [delta1, delta2, delta3]

# Next-Nearest-Neighbor Vectors (A → A)
b1 = np.array([np.sqrt(3) * a, 0])
b2 = np.array([-np.sqrt(3)/2 * a, -1.5 * a])
b3 = np.array([-np.sqrt(3)/2 * a, 1.5 * a])
b_vecs = [b1, b2, b3]

# Geometry for finite system (ribbon)
N_x = 40   # Number of unit cells in x (width)
N_y = 20   # Number of sites in y (periodic)
N_sites = 2 * N_x * N_y

Nk = 200       # Number of k-points
k_vals = np.linspace(-np.pi, np.pi, Nk)
k_y = 0.0
# Add BZ boundary
bz = np.array([
    [2*np.pi/np.sqrt(3), 0],
    [np.pi/np.sqrt(3), np.pi],
    [-np.pi/np.sqrt(3), np.pi],
    [-2*np.pi/np.sqrt(3), 0],
    [-np.pi/np.sqrt(3), -np.pi],
    [np.pi/np.sqrt(3), -np.pi],
    [2*np.pi/np.sqrt(3), 0]
])
```

```python
# Define d_x(k), d_y(k), d_z(k)
def d_x(kx, ky):
    """Nearest-neighbor hopping: real part"""
    total = 0.0
    k = np.array([kx, ky])
    for d in deltas:
        total += np.cos(np.dot(k, d))
    return -t1 * total

def d_y(kx, ky):
    """Nearest-neighbor hopping: imaginary/antisymmetric part"""
    total = 0.0
    k = np.array([kx, ky])
    for d in deltas:
        total += np.sin(np.dot(k, d))
    return -t1 * total

def d_z(kx, ky):
    """Staggered potential + Haldane flux term"""
    total = 0.0
    k = np.array([kx, ky])
    for b in b_vecs:
        total += np.sin(np.dot(k, b))
    return t0 - 2 * t2 * np.sin(phi) * total
```

```python
# Energy dispersion E±(k) = ±|d(k)|
def energy_plus(kx, ky):
    dx = d_x(kx, ky)
    dy = d_y(kx, ky)
    dz = d_z(kx, ky)
    return np.sqrt(dx**2 + dy**2 + dz**2)

def energy_minus(kx, ky):
    return -energy_plus(kx, ky)
```

```python
# Create k-grid
nk = 200
k_vals = np.linspace(-2*np.pi/np.sqrt(3), 2*np.pi/np.sqrt(3), nk)
KX, KY = np.meshgrid(k_vals, k_vals)

# Evaluate energy on grid
EK_plus = np.array([[energy_plus(kx, ky) for kx in k_vals] for ky in k_vals])
EK_minus = -EK_plus
```

```python
# High-symmetry points (for honeycomb lattice, e.g., graphene)
Gamma = np.array([0, 0])
K = np.array([4 * np.pi / (3 * np.sqrt(3) * a), 0])
M = np.array([np.pi / (np.sqrt(3) * a), np.pi / a])

# Path: Γ → K → M → Γ
path_points = [Gamma, K, M, Gamma]
labels = [r'$\Gamma$', r'$K$', r'$M$', r'$\Gamma$']

def interpolate_path(points, n_per_segment=100):
    """
    Interpolate k-points along a path.
    Returns: kx_vals, ky_vals, and x_ticks for plotting.
    """
    kx_vals, ky_vals = [], []
    x_ticks = [0]  # Starting point at 0
    cumulative = 0

    for i in range(len(points) - 1):
        start = points[i]
        end = points[i+1]
        kxs = np.linspace(start[0], end[0], n_per_segment)
        kys = np.linspace(start[1], end[1], n_per_segment)
        
        kx_vals.extend(kxs)
        ky_vals.extend(kys)
        
        cumulative += n_per_segment
        x_ticks.append(cumulative)
    
    return np.array(kx_vals), np.array(ky_vals), x_ticks

# Generate k-path
kx_vals, ky_vals, x_ticks = interpolate_path(path_points, n_per_segment=100)

# Evaluate energies
E_plus = np.array([energy_plus(kx, ky) for kx, ky in zip(kx_vals, ky_vals)])
E_minus = np.array([energy_minus(kx, ky) for kx, ky in zip(kx_vals, ky_vals)])
```

```python
plt.figure(figsize=(8, 6))

# Plot bands
plt.plot(E_plus, color='black', label=r'$E_+(k)$', linewidth=2)
plt.plot(E_minus, color='black', label=r'$E_-(k)$', linewidth=2)

# Format x-axis
plt.xticks(ticks=x_ticks, labels=labels)
for x in x_ticks:
    plt.axvline(x=x, color='gray', linestyle='--', alpha=0.7)
    
plt.ylabel('Energy $E(k)$', fontsize=12)
plt.title('Haldane Model Band Structure', fontsize=14)
plt.grid(True, alpha=0.3)

# Optional: set y-limits for clarity
plt.ylim(-3, 3)

plt.tight_layout()
plt.show()
```

```python
def build_haldane_ribbon(k_y, t1=1.0, t2=0.1, t0=0.3, phi=np.pi/2, N_x=30, N_y=1):
    """
    Build Haldane model Hamiltonian for a ribbon:
    - Open along x (finite N_x)
    - Momentum k_y along y
    """
    def index(x, sublattice):
        return 2*x + sublattice  # y collapsed into momentum

    N_sites = 2 * N_x
    H = np.zeros((N_sites, N_sites), dtype=complex)

    for x in range(N_x):
        iA = index(x, 0)
        iB = index(x, 1)

        # On-site staggered potential
        H[iA, iA] += t0 / 2
        H[iB, iB] -= t0 / 2

        # === NN Hopping (A→B) ===
        # δ1 = (0,0)
        H[iA, iB] += -t1
        H[iB, iA] += -t1

        # δ2 = (-1,0)
        if x > 0:
            jB = index(x-1, 1)
            H[iA, jB] += -t1 * np.exp(-1j * k_y)  # phase from k_y
            H[jB, iA] += -t1 * np.exp(1j * k_y)

        # δ3 = (-1,1)
        if x > 0:
            jB = index(x-1, 1)
            H[iA, jB] += -t1
            H[jB, iA] += -t1

        # === NNN Hopping (complex phases) ===
        # A→A hoppings
        if x < N_x-1:
            jA = index(x+1, 0)
            H[iA, jA] += -t2 * np.exp(1j * phi)
            H[jA, iA] += -t2 * np.exp(-1j * phi)
        if x > 0:
            jA = index(x-1, 0)
            H[iA, jA] += -t2 * np.exp(-1j * phi)
            H[jA, iA] += -t2 * np.exp(1j * phi)

        # B→B hoppings
        if x < N_x-1:
            jB = index(x+1, 1)
            H[iB, jB] += -t2 * np.exp(1j * phi)
            H[jB, iB] += -t2 * np.exp(-1j * phi)
        if x > 0:
            jB = index(x-1, 1)
            H[iB, jB] += -t2 * np.exp(-1j * phi)
            H[jB, iB] += -t2 * np.exp(1j * phi)

    return H
```

```python
# Build Hamiltonian with default parameters
H = build_haldane_ribbon(k_y, t1=t1, t2=t2, t0=0.3, phi=phi, N_x=N_x)

# Diagonalize
eigenvalues, eigenvectors = np.linalg.eigh(H)

# Sort by energy
idx = np.argsort(eigenvalues)
eigenvalues = eigenvalues[idx]

# Plot: Eigenvalues vs Index
plt.figure(figsize=(10, 6))
plt.plot(eigenvalues, '-b', markersize=3)

plt.xlabel('Site Index')
plt.ylabel('Energy $E$')
plt.title(f'Haldane Model: Eigenvalues vs Index\n'
          f'$N_x={N_x}, N_y={N_y}, t_1={t1}, t_2={t2}, t_0={t0}, \\phi=\\pi/2$')
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```

```python
# Compute spectrum E(k_y)
bands = []

for ky in k_vals:
    H = build_haldane_ribbon(k_y=ky, t1=t1, t2=t2, t0=0.3, phi=phi, N_x=N_x)
    eigvals = np.linalg.eigvalsh(H)
    bands.append(eigvals)

bands = np.array(bands)

# Plot Band Structure
plt.figure(figsize=(8, 6))

# Plot all bands
for n in range(bands.shape[1]):
    plt.plot(k_vals, bands[:, n], 'darkblue','b-', lw=0.6, alpha=0.8)

# Highlight Fermi level
plt.axhline(0, color='k', linestyle='--', alpha=0.7, label='Fermi level')

# Formatting
plt.xlabel(r'$k_y$', fontsize=12)
plt.ylabel('Energy $E$', fontsize=12)
plt.title(f'Haldane Ribbon: $E$ vs $k_y$\n'
          f'$t_1={t1}, t_2={t2}, t_0={t0}, \\phi=\\pi/2, N_x={N_x}$', fontsize=10)
plt.xlim(-np.pi, np.pi)
plt.legend()
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```

## Chern Phase Diagram:

```python
# Pauli matrices (needed for H_k)
sigma_x = np.array([[0, 1], [1, 0]], dtype=complex)
sigma_y = np.array([[0, -1j], [1j, 0]], dtype=complex)
sigma_z = np.array([[1, 0], [0, -1]], dtype=complex)

def H_k(kx, ky, t1=1.0, t2=0.3, t0=0.2, phi=np.pi/2):
    """Construct 2x2 Hamiltonian using your d_x, d_y, d_z"""
    dx = d_x(kx, ky)
    dy = d_y(kx, ky)
    dz = d_z(kx, ky)
    return dx * sigma_x + dy * sigma_y + dz * sigma_z
```

```python
# Compute Chern Number Numerically (Fukui-Hatsugai Method)

def compute_chern(t2_val, t0_val, phi_val, nx=40, ny=40):
    """
    Compute Chern number for given t2, t0, phi
    Uses your d_x, d_y, d_z with updated t0, t2, phi
    """
    # Temp functions with current parameters
    def d_x_temp(kx, ky):
        total = 0.0
        k = np.array([kx, ky])
        for d in deltas:
            total += np.cos(np.dot(k, d))
        return -t1 * total

    def d_y_temp(kx, ky):
        total = 0.0
        k = np.array([kx, ky])
        for d in deltas:
            total += np.sin(np.dot(k, d))
        return -t1 * total

    def d_z_temp(kx, ky):
        total = 0.0
        k = np.array([kx, ky])
        for b in b_vecs:
            total += np.sin(np.dot(k, b))
        return t0_val - 2 * t2_val * np.sin(phi_val) * total

    # Build H_k with current params
    def H_k_temp(kx, ky):
        dx = d_x_temp(kx, ky)
        dy = d_y_temp(kx, ky)
        dz = d_z_temp(kx, ky)
        return dx * sigma_x + dy * sigma_y + dz * sigma_z

    # Grid in k-space
    kx_vals = np.linspace(-2*np.pi/np.sqrt(3), 2*np.pi/np.sqrt(3), nx)
    ky_vals = np.linspace(-np.pi, np.pi, ny)

    # Link variables
    Ux = np.ones((nx, ny), dtype=complex)
    Uy = np.ones((nx, ny), dtype=complex)

    for i in range(nx):
        for j in range(ny):
            kx, ky = kx_vals[i], ky_vals[j]
            H = H_k_temp(kx, ky)
            evals, evecs = np.linalg.eigh(H)
            u = evecs[:, 0]  # Occupied band (lower energy)

            # Ux link: k → k + dk_x
            kx_next = kx_vals[(i+1) % nx]
            H_next_x = H_k_temp(kx_next, ky)
            _, evecs_next = np.linalg.eigh(H_next_x)
            u_next_x = evecs_next[:, 0]
            Ux[i,j] = np.vdot(u, u_next_x) / abs(np.vdot(u, u_next_x))

            # Uy link: k → k + dk_y
            ky_next = ky_vals[(j+1) % ny]
            H_next_y = H_k_temp(kx, ky_next)
            _, evecs_next = np.linalg.eigh(H_next_y)
            u_next_y = evecs_next[:, 0]
            Uy[i,j] = np.vdot(u, u_next_y) / abs(np.vdot(u, u_next_y))

    # Compute Berry curvature flux per plaquette
    F = np.zeros((nx, ny))
    for i in range(nx):
        for j in range(ny):
            prod = Ux[i,j] * Uy[(i+1)%nx, j] * np.conj(Ux[i, (j+1)%ny]) * np.conj(Uy[i,j])
            F[i,j] = np.angle(prod)

    chern = np.sum(F) / (2*np.pi)
    return np.round(chern).astype(int)
```

```python
# Grid resolution
n_phi = 50
n_ratio = 40
t2_fixed = 0.3

phi_vals = np.linspace(0, 2*np.pi, n_phi)
ratio_vals = np.linspace(-6, 6, n_ratio)

chern_map = np.zeros((n_ratio, n_phi), dtype=int)

print("Computing phase diagram in parallel...")

def compute_row(i, ratio):
    row = []
    for phi in phi_vals:
        t0_val = ratio * t2_fixed
        C = compute_chern(t2_val=t2_fixed, t0_val=t0_val, phi_val=phi)
        row.append(C)
    return i, row

# Parallel loop (use n_jobs=-1 for all cores)
results = Parallel(n_jobs=-1, backend="loky")(delayed(compute_row)(i, ratio) 
                                              for i, ratio in enumerate(ratio_vals))

# Collect results
for i, row in results:
    chern_map[i, :] = row

print("Phase diagram computed.")
```

```python
# Plot Phase Diagram

plt.figure(figsize=(10, 6))

im = plt.imshow(chern_map, aspect='auto',
                origin='lower',
                extent=[0, 2*np.pi, -6, 6],
                cmap='RdYlBu_r',
                interpolation='nearest')

plt.xlabel(r'$\phi$')
plt.ylabel(r'$t_0 / t_2$')
plt.title('Haldane Model Phase Diagram\nChern Number $\\mathcal{C}$')

plt.xticks([0, np.pi/2, np.pi, 3*np.pi/2, 2*np.pi],
           [r'$0$', r'$\pi/2$', r'$\pi$', r'$3\pi/2$', r'$2\pi$'])

cbar = plt.colorbar(im, label='Chern Number $\\mathcal{C}$')
cbar.set_ticks([-1, 0, 1])
cbar.set_ticklabels(['-1', '0', '1'])

plt.tight_layout()
plt.show()
```
